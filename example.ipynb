```python
import json
import GenderedLanguageAnalysisTool as glat
from pathlib import Path
```

```python
# Opening gendered terms example file 
p = Path('gendered_terms_example_article.json')
term_ex = json.loads(p.read_text())

# This file contains the example article text as well as identifying information that can be used to cite the article
# The Associated Press. (2024, October 1). John Amos, star of Good Times and Roots, dead at 84. CBC. https://www.cbc.ca/news/entertainment/john-amos-good-times-roots-obituary-1.7339138
```

```python
# Initialize the GLAT instance
tool_ex1 = glat.GenderedLanguageAnalysisTool()

# Analyze the text of the example article
tool_ex1.analyse_text(term_ex['text'])

# Print out the gender analysis details
tool_ex1.display_info()

# Display visualization of the text
tool_ex1.visualize_doc()
```
